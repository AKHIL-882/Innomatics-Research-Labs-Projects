**1.1 Import the required libraries**

```python
import numpy as np
import pandas as pd
```

**1.2 Reading the datasets**

```python
dataset_data = pd.read_csv('u.data',sep='\t',names=['user_id','item_id','rating','timestamp'])


dataset_item = pd.read_csv("u.item", sep="|", encoding="iso-8859-1",names=["item_id","item_name","date","unknown1"
"website","rat1","rat2","rat3","rat4","rat5","rat6","rat7","rat8","rat9","rat10","rat11","rat12","rat13",
"rat14","rat15","rat16","rat17","rat18","rat19","rat20"])
```

**Observation**: We can observe that there is one columns which has all null values**(i.e, unknown1website) column**

**1.7 Slicing the dataset_item**

```python
dataset_item_new = dataset_item.iloc[:,0:2]
dataset_item_new.head()
```

**Observation**: We have sliced the dataset for further processing

**1.8 Merging the dataframes**

```python
dataset_new = dataset_data.merge(dataset_item_new,on="item_id")
dataset_new.drop(['timestamp'],inplace=True,axis=1)
dataset_new.head()
```

**Observation**: We have merged the two dataframes based on the **item_id**

**1.9 Pivot table**

```python
data_table = pd.pivot_table(dataset_new,values='rating',columns='item_name',index='user_id')
data_table.head()
```

**2.0 Building the recomendations**

**2.1.1 User Engagement**

```python
movie_name = input("Enter the movie name: ")
```

**Observation**: User will enter the specific movie to which the result would be matched

**2.1.2 Finding the recomendation of the movie requested**

```python
print(" Here are the list of movies which are matched to your search")
print(data_table.corr()[movie_name].sort_values(ascending=False).iloc[:20])
```

**Observation**: We are displaying the top 20 movies for the user which is slight related to the movie specified by the user
