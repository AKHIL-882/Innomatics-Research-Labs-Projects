**2.6 Dropping the null values**

```python
dataset_triplets.dropna(axis='columns')
dataset_triplets.dropna(how='all')
dataset_triplets.dropna(subset=['user_id', 'song_id',"song_count"])
```

**2.7 Grouping the dataset based on userid**

```python
dataset_triplets_new= pd.DataFrame(dataset_triplets.groupby('user_id')['song_count'].sum()) 
dataset_triplets_new
```

**2.8 Merging the triplets dataset based on user_id**

```python
popularity_dataset_triplets=pd.merge(dataset_triplets_new,dataset_triplets, how="left", on=["user_id"]) 
popularity_dataset_triplets
```

**2.9 Merging the tracks dataset based on the song_id**

```python
popularity_dataset_tracks=pd.merge(popularity_dataset_triplets,dataset_unique_tracks,how="left",on=["song_id"])
popularity_dataset_tracks
```

**3.0 Creating a new columns for most popular songs**

```python
popularity_dataset_tracks['most_popularity'] = popularity_dataset_tracks['song_count_x'].rank(method='max')
popularity_dataset_tracks
```

**3.1 Creating a new columns for least popular songs**

```python
popularity_dataset_tracks['least_popularity'] = popularity_dataset_tracks['song_count_x'].rank(method='min')
popularity_dataset_tracks
```

**3.2 Calculating the percentage of ranks song**

```python
popularity_dataset_tracks['Percentage_Rank'] = popularity_dataset_tracks['song_count_x'].rank(pct=True)
popularity_dataset_tracks
```

**3.3 Displaying the songs count in decending order**

```python
popularity_dataset_tracks=popularity_dataset_tracks.sort_values(by='song_count_x', ascending=False)
popularity_dataset_tracks
```

**3.4 Building the model**

```python
artist=popularity_dataset_tracks.groupby(by='singer_name')
artist_new=artist['user_id'].nunique()
artist_new.sort_values(inplace=True,ascending=False)

user_id_input=input("Enter the user_id:")
print("Artist based popularity system")
recommendation=artist_new.head(10)
print("The recommendation:\n",recommendation)
```
